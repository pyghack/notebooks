```
import pandas as pd
import numpy as np
import plotly
```

```
df = pd.read_csv("https://raw.githubusercontent.com/danielferriss/PygArtHackTheater2019/master/manipulation/sums.csv", delimiter=',')
df = df.sort_values(by=['Quantity'])
```

```
import plotly.graph_objects as go

fig = go.Figure()


fig.add_trace(go.Bar(name='Ticket', y=df['Events'], x=df['Price'], orientation='h'))
fig.add_trace(go.Bar(name='Merchandise', y=df['Events'], x=df['Merchandise_Sales'], orientation='h'))
fig.add_trace(go.Bar(name='Concession', y=df['Events'], x=df['Concession_Sales'], orientation='h'))

fig.update_layout(
    autosize=True,
    height=8000,
    margin=go.layout.Margin(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
    xaxis=dict(
        title="Sales",
        titlefont=dict(
            color="#1f77b4"
        ),
        tickfont=dict(
            color="#1f77b4"
        ),
        side='top'
    )
)

# Change the bar mode
fig.update_layout(barmode='stack')
fig.show()
```

```
import plotly.offline
from google.colab import files

file_name = 'sales_by_showings.html'
plotly.offline.plot(fig, filename=file_name)
```
